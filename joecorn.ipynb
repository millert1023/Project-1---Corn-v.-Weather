```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
iowa = "../Project-1---Corn-v.-Weather/Iowa_weather_2016to2018.csv"
iowa_df = pd.read_csv(iowa)
iowa_df.head()
```

```python
indiana = "../Project-1---Corn-v.-Weather/Indiana_weather_2016to2018.csv"
indiana_df = pd.read_csv(indiana)
indiana_df.head()
```

```python
illinois = "../Project-1---Corn-v.-Weather/Illinois_weather_2016to2018.csv"
illinois_df = pd.read_csv(illinois)
illinois_df.head()
```

```python
nebraska = "../Project-1---Corn-v.-Weather/Nebraska_weather_2016to2018.csv"
nebraska_df = pd.read_csv(nebraska)
nebraska_df.head()
```

```python
state_combined = pd.concat([indiana_df, iowa_df, nebraska_df, illinois_df])
```

```python
state_combined.count()
```

```python
state_combined_df = state_combined[["STATE", "DATE", "ELEVATION", "PRCP", "SNOW", "TAVG", "TMAX", "TMIN"]]
state_combined_df.head()
```

```python
state_combined_df = state_combined_df.dropna()
```

```python
state_combined_df.head()
```

```python
state_combined_df.count()
```

```python
state_weather = state_combined_df.groupby(['STATE', 'DATE']).mean()
```

```python
state_weather.head()
```

```python
corn_price = "../Project-1---Corn-v.-Weather/Corn_price_2016to2018.csv"
corn_price_df = pd.read_csv(corn_price)
corn_price_df.head()
```

```python
corn_price_pd = corn_price_df[["STATE","DATE","PRICE RECEIVED -- $ / BU"]]
corn_price_pd.head()
```

```python
corn_price_pd = corn_price_pd.dropna()
```

```python
corn_price_pd.head(20)
```

```python
corn_ear = "../Project-1---Corn-v.-Weather/Corn_ear_cnt_2016to2018.csv"
corn_ear_df = pd.read_csv(corn_ear)
corn_ear_df.head()
```

```python
corn_ear_df = corn_ear_df.dropna()
corn_ear_df.head()
```

```python
corn_pr_ear_df = pd.merge(corn_price_pd,corn_ear_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_pr_ear_df.head()
```

```python
corn_prod_price = "../Project-1---Corn-v.-Weather/Corn_production by money_2016to2018.csv"
corn_prod_price_df = pd.read_csv(corn_prod_price)
corn_prod_price_df.head()
```

```python
corn_prod_price_df = corn_prod_price_df.dropna()
corn_prod_price_df
```

```python
corn_pro_df = pd.merge(corn_pr_ear_df,corn_prod_price_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_pro_df.head()
```

```python
corn_prod = "../Project-1---Corn-v.-Weather/Corn_production_2016to2018.csv"
corn_prod_df = pd.read_csv(corn_prod)
corn_prod_df.head()
```

```python
corn_prod_df = corn_prod_df.dropna()
corn_prod_df.head()
```

```python
corn_production_df = pd.merge(corn_pro_df,corn_prod_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_pro_df.head()
```

```python
corn_df = pd.merge(state_weather,corn_production_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_df.head(20)
```
