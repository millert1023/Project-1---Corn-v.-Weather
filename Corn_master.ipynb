```python
#Import Dependencies
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import linregress
```

```python
#Import CSV
iowa = "../Project-1---Corn-v.-Weather/Iowa_weather_2016to2018.csv"
iowa_df = pd.read_csv(iowa)
iowa_df.head()
```

```python
#Import CSV
indiana = "../Project-1---Corn-v.-Weather/Indiana_weather_2016to2018.csv"
indiana_df = pd.read_csv(indiana)
indiana_df.head()
```

```python
#Import CSV
illinois = "../Project-1---Corn-v.-Weather/Illinois_weather_2016to2018.csv"
illinois_df = pd.read_csv(illinois)
illinois_df.head()
```

```python
#Import CSV
nebraska = "../Project-1---Corn-v.-Weather/Nebraska_weather_2016to2018.csv"
nebraska_df = pd.read_csv(nebraska)
nebraska_df.head()
```

```python
#Merge to single dataset 
state_combined = pd.concat([indiana_df, iowa_df, nebraska_df, illinois_df])
```

```python
state_combined.count()
```

```python
#Remove columns
state_combined_df = state_combined[["STATE", "DATE", "ELEVATION", "PRCP", "SNOW", "TAVG", "TMAX", "TMIN"]]
state_combined_df.head()
```

```python
#Drop NAs
state_combined_df = state_combined_df.dropna()
```

```python
state_combined_df.head()
```

```python
state_combined_df.count()
```

```python
#Group by State and Date then find the averages
state_weather = state_combined_df.groupby(['STATE', 'DATE']).mean()
```

```python
state_weather.head()
```

```python
#Import CSV
corn_price = "../Project-1---Corn-v.-Weather/Corn_price_2016to2018.csv"
corn_price_df = pd.read_csv(corn_price)
corn_price_df.head()
```

```python
#Remove columns
corn_price_pd = corn_price_df[["STATE","DATE","PRICE RECEIVED -- $ / BU"]]
corn_price_pd.head()
```

```python
#Drop NAs
corn_price_pd = corn_price_pd.dropna()
```

```python
corn_price_pd.head(20)
```

```python
#Import CSV
corn_ear = "../Project-1---Corn-v.-Weather/Corn_ear_cnt_2016to2018.csv"
corn_ear_df = pd.read_csv(corn_ear)
corn_ear_df.head()
```

```python
#Drop NAs
corn_ear_df = corn_ear_df.dropna()
corn_ear_df.head()
```

```python
#Merge DataFrames -- Production/Price
corn_pr_ear_df = pd.merge(corn_price_pd,corn_ear_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_pr_ear_df.head()
```

```python
#Import CSV
corn_prod_price = "../Project-1---Corn-v.-Weather/Corn_production by money_2016to2018.csv"
corn_prod_price_df = pd.read_csv(corn_prod_price)
corn_prod_price_df.head()
```

```python
#Drop NAs
corn_prod_price_df = corn_prod_price_df.dropna()
corn_prod_price_df
```

```python
#Merge DataFrames -- Production/Price

corn_pro_df = pd.merge(corn_pr_ear_df,corn_prod_price_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_pro_df.head()
```

```python
#Import CSV
corn_prod = "../Project-1---Corn-v.-Weather/Corn_production_2016to2018.csv"
corn_prod_df = pd.read_csv(corn_prod)
corn_prod_df.head()
```

```python
#Drop NAs
corn_prod_df = corn_prod_df.dropna()
corn_prod_df.head()
```

```python
#Merge DataFrames -- Production/Price
corn_production_df = pd.merge(corn_pro_df,corn_prod_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_production_df.head()
```

```python
#Merge DataFrames -- Weather and Production/Price
corn_df = pd.merge(state_weather,corn_production_df,  how='left', left_on=['STATE', 'DATE'], right_on=["STATE", "DATE"])
corn_df.head(20)
```

```python
#Convert String to Float
corn_df['PRODUCTION, MEASURED IN BU'] = (corn_df['PRODUCTION, MEASURED IN BU'].apply(lambda x: float(x.split()[0].replace(',', '')))/(1000000000)).round(2)
```

```python
#Convert String to Float
corn_df['EAR COUNT - PER ACRE'] = (corn_df['EAR COUNT - PER ACRE'].apply(lambda x: float(x.split()[0].replace(',', '')))).round(0)
```

```python
#Convert String to Float
corn_df['PRODUCTION, MEASURED IN $'] = (corn_df['PRODUCTION, MEASURED IN $'].apply(lambda x: float(x.split()[0].replace(',', '')))/(1000000000)).round(2)
```

```python
corn_df.head()
```

```python
corn_df.dtypes
```

```python
# Minor Data Munging to Re-Format the Data Frames
change_corn_df = corn_df.pivot(index='DATE', columns='STATE', values='EAR COUNT - PER ACRE')
change_corn_df.reset_index("DATE", inplace=True)
change_corn_df
```

```python
# Generate the Plot
plt.style.use('seaborn-darkgrid')
plt.plot(change_corn_df ["DATE"], change_corn_df["ILLINOIS"],marker="", color="RED",
                 linewidth=1, alpha=0.9, label="Illinois")
plt.plot(change_corn_df ["DATE"], change_corn_df["INDIANA"],marker="", color="blue",
                 linewidth=1, alpha=0.9, label="Indiana")
plt.plot(change_corn_df ["DATE"], change_corn_df["NEBRASKA"],marker="", color="green",
                 linewidth=1, alpha=0.9, label="Nebraska")
plt.plot(change_corn_df ["DATE"], change_corn_df["IOWA"],marker="", color="black",
                 linewidth=1, alpha=0.9, label="Iowa")
    
plt.legend(loc="best")
plt.title("Corn Ear Count Over the Years", fontsize=16, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Ear Count per Acre")
plt.xticks(change_corn_df["DATE"])

# Save the Figure
plt.savefig("../Project-1---Corn-v.-Weather/Ear_ct_year.png")


# Show the Figure
plt.show()
```

```python
# Generate the Plot
x_values = corn_df['TAVG']
y_values = corn_df['PRODUCTION, MEASURED IN BU']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
# line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
# plt.annotate(line_eq,(0,70),fontsize=15,color="red")
plt.xlabel("Avg Temperature (F)")
plt.ylabel("Production measure in Bushels (in billions)")
print(f"The r-squared is: {rvalue}")
plt.title("Corn Production by Bushels vs Temperature", fontsize=16, fontweight='bold')

# Save the Figure
plt.savefig("../Project-1---Corn-v.-Weather/Production_temp.png")

# Show the Figure
plt.show()
```
